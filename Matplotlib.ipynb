```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(4,15,100)
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(5,3))
plt.plot(x, y, label="Sin Wave", linestyle="-", linewidth=2, color="#00ff00")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Sin Wave")
plt.grid(True)
plt.show()
```

```python
category = ["A", "B", "C", "D", "E"]
value = [33, 54, 42, 51, 29]
```

```python
plt.figure(figsize=(5,3))
plt.bar(category, value, alpha=0.6, edgecolor="black",
        color=["#234269","#234321", "#234290", "#234310", "#234250"])
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Comparing Categores")
plt.grid(True)
plt.show()
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
color = np.random.rand(100)
size = 1000 * np.random.rand(100)
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x, y, c=color, s=size, alpha=0.5, cmap="plasma")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Scatter Plot")
plt.colorbar(label="Color Intensity")
plt.grid(True)
plt.show()
```

```python
dice = np.array([1, 2, 3, 4, 5, 6]) #Dice
sample = np.random.choice(dice, 4) # Rplling dice 4 time in a row and getting values
print("Output of 4 samples", sample)
print("Mean of aboue 4 samples", sample.mean())
```

```python
samples_1000 = [np.random.choice(dice, 4).mean() for i in range(1000)]
samples_1000
```

```python
plt.figure(figsize=(5,3))
plt.hist(samples_1000, alpha=0.7, edgecolor="black", bins=100)
plt.show()
```

```python
data = np.random.randn(1000)

plt.figure(figsize=(5,3))
plt.hist(data, alpha=0.7, edgecolor="black", bins=100, histtype="stepfilled")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.grid(True)
plt.show()
```

```python
sizes = np.random.randint(35, 89, 4)
labels = ["East", "West", "North", "South"]
plt.figure(figsize=(5,4))
plt.pie(sizes, labels=labels, colors=["gold", "lightgreen", "cyan","lightcoral"],
       autopct="%1.1f%%", explode=[0, 0.2, 0, 0])
plt.title("Piechar")
plt.show()
```

```python
data = np.random.normal(loc=0, scale=1, size=(100,))

plt.figure(figsize=(5,3))
plt.boxplot(data, patch_artist=True, vert=False)
plt.title("Boxplot")
plt.show()
```
