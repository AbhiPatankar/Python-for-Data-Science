```python
import pandas as pd
import numpy as np
```

```python
arr = np.random.randint(40,99,24).reshape(6,4)
arr
```

```python
df = pd.DataFrame(arr, columns=["Math", "English", "Physics", "Sport"]) # Creating dataframe
df["Std_Name"] = ["Sachin", "Rohit", "Shubhaman", "Virat", "Ravindra", "Rinku"] # Adding column
cols = ["Std_Name","Math", "English", "Physics", "Sport"] # Changing position
df = df[cols]
df.head()
```

```python
df1 = df[["Std_Name", "Math", "English",]]
df2 = df[["Std_Name", "Physics", "Sport"]]
```

```python
df1.head(2), df2.head(2) #this make the output tuple
```

```python
type((df1.head(2), df2.head(2)))
```

```python
df1.merge(df2, on="Std_Name", how= "outer")
```

```python
df1 = df1[df1["Std_Name"]!="Sachin"]
```

```python
df1
```

```python
df2 = df2[df2["Std_Name"]!="Rinku"]
df2
```

```python
df1.merge(df2, on="Std_Name", how="outer")
```
