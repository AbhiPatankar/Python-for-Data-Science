```python
import seaborn as sns
import matplotlib.pyplot as plt
#from matplotlib import pyplot as plt
```

```python
weight = [62, 64, 69, 75, 66, 68, 65, 71, 76, 73]
height = [120, 136, 148, 175, 137, 165, 154, 172, 200 , 187]
```

```python
%matplotlib inline
plt.figure(figsize=(4,3.5))
plt.xlabel("Weight")
plt.ylabel("Height")
plt.title("Height vs Weight")
sns.scatterplot(x = height, y=weight)
plt.show()
```

```python
gender = ["Female", "Female", "Female", "Female",
          "Male", "Male", "Male", "Male","Male", "Male",]
plt.figure(figsize=(4.5,3))
sns.countplot(x=gender, hue=gender,legend=["Female", "Male"], fill=False)
plt.show()
```

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel("SampleSuperstore.xlsx")
df.head()
```

```python
sns.countplot(data=df, x="Ship Mode", hue="Segment")
plt.show()
```

```python
df.columns
```

```python
sns.scatterplot(data=df, x="Quantity", y="Sales", hue="Category")
plt.show()
```

```python
plt.figure(figsize=(19,5))
sns.histplot(data=df, x="Sub-Category")
plt.xticks(rotation=60)
plt.show()
```

```python
sns.scatterplot(data=df, x="Quantity", y="Sales", hue="Category", 
                hue_order=["Technology", "Office Supplies", "Furniture"],
               palette={"Technology":"red", "Office Supplies":"blue", "Furniture":"black" })
plt.show()
```

```python
sns.scatterplot(data=df, x="Quantity", y="Sales", hue="Category", 
                hue_order=["Technology", "Office Supplies", "Furniture"],
                palette={"Technology": "#808080","Office Supplies":"#00ff00", "Furniture":"#230023" })
plt.show()
```
